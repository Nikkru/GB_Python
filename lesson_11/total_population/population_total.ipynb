```python
# смотрим страны с популяцией полеее 40 млн за 2005 год
# Знак & означает 'and', а знак | 'or'
df[(df['population'] > 40000000) & (df['year'] == 2005)].sort_values('population', ascending=False)
```

```python
# смотрим самые малонаселенные страны в 1995 году
min = df['population'].min()
min
df[(df['population'] < min*20) & (df['year'] == 1995)]
```

```python
# смотрим самые заселенные страны в 1995 году
max = df['population'].max()
df[(df['population'] > max*0.7) & (df['year'] == 1995)]
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import seaborn as sns
from IPython.display import set_matplotlib_formats
```

```python
# фильтруем данные, отбиравя значения по самым заселенным странам в 1990 году, выполняем сортировку по величине популяции
max = (df['year'] == 1990).max()
df_1990 = df[df['year']==1990]
max_1990 = df_1990['population'].max()
df_1990_max = df_1990[(df_1990['population']>max_1990*0.04)].sort_values('population', ascending=False)
df_1990_max
```

Для анализа динамики и наблюдения тенденций выбрана целевая группа из 23 самых крупных по населению стран. В представленном ниже графике можно видеть, что более половины суммы населения рассматриваемой группы принадлежит двум страннам: Китаю и Индии. Разрыв с население других стран здесь так велик, что можно рассматривать эти траны как пару, с одной тенденцией. 

```python
# строим график, визуализирующий объем населения стран из выбранной группы для сравнения
x = df_1990_max['country'].tolist()
y = df_1990_max['population'].tolist()
fig = plt.figure(figsize=(15,15))
plt.pie(y, labels=(x))
plt.show()
```

На графике с кривыми прогресса наслеления хорошо видно, что вплоть до 90-х годов линии Китая и Индии шли круто вверх параллельно друг другу, но в 90-х темпы роста населения в Китае заметно замедляются, в верхней части графика кривые уже теряют свой паралельность и начинают сближение. Индийская продолжает свой неуклонных рост до настоящего времени, китайская меняет угол на более пологий, и за последние 30 лет он уже не меняется, что говорит об перманентности произошедших изменений и устойчивости настоящих тенденций.
